```python
# Axel Guzman NL = 23
# Fecha 06/05/2026

# UIII Act 3 Listas con Funciones

# Sistema de Dentista en Python


# Lista donde se guardarán los pacientes
pacientes = []

# --------------------------------------------------
# Función agregar paciente
def agregar_paciente():

    nombre = input("Ingresa el nombre del paciente: ")
    edad = input("Ingresa la edad del paciente: ")
    problema = input("Ingresa el problema dental: ")

    paciente = {
        "nombre": nombre,
        "edad": edad,
        "problema": problema
    }

    pacientes.append(paciente)

    print("Paciente agregado correctamente.")

# --------------------------------------------------
# Función leer paciente
def leer_paciente():

    nombre = input("Ingresa el nombre del paciente a leer: ")

    for paciente in pacientes:

        if paciente["nombre"] == nombre:

            print("Nombre:", paciente["nombre"])
            print("Edad:", paciente["edad"])
            print("Problema:", paciente["problema"])

            return

    print("Paciente no encontrado.")

# --------------------------------------------------
# Función listar pacientes
def listar_pacientes():

    if len(pacientes) == 0:

        print("No hay pacientes registrados.")

    else:

        print("\nLista de pacientes:")

        for paciente in pacientes:

            print("----------------------")
            print("Nombre:", paciente["nombre"])
            print("Edad:", paciente["edad"])
            print("Problema:", paciente["problema"])

# --------------------------------------------------
# Función buscar paciente
def buscar_paciente():

    nombre = input("Ingresa el nombre del paciente a buscar: ")

    encontrado = False

    for paciente in pacientes:

        if paciente["nombre"] == nombre:

            print("Paciente encontrado")
            print("Nombre:", paciente["nombre"])
            print("Edad:", paciente["edad"])
            print("Problema:", paciente["problema"])

            encontrado = True

    if encontrado == False:

        print("Paciente no encontrado.")

# --------------------------------------------------
# Función modificar paciente
def modificar_paciente():

    nombre = input("Ingresa el nombre del paciente a modificar: ")

    for paciente in pacientes:

        if paciente["nombre"] == nombre:

            paciente["nombre"] = input("Nuevo nombre: ")
            paciente["edad"] = input("Nueva edad: ")
            paciente["problema"] = input("Nuevo problema dental: ")

            print("Paciente modificado correctamente.")

            return

    print("Paciente no encontrado.")

# --------------------------------------------------
# Función borrar paciente
def borrar_paciente():

    nombre = input("Ingresa el nombre del paciente a borrar: ")

    for paciente in pacientes:

        if paciente["nombre"] == nombre:

            pacientes.remove(paciente)

            print("Paciente eliminado correctamente.")

            return

    print("Paciente no encontrado.")

# --------------------------------------------------
# Función menú
def menu():

    while True:

        print("\n===== MENU DENTISTA =====")
        print("1. Agregar paciente")
        print("2. Leer paciente")
        print("3. Listar pacientes")
        print("4. Buscar paciente")
        print("5. Modificar paciente")
        print("6. Borrar paciente")
        print("7. Salir")

        opcion = int(input("Selecciona una opción: "))

        if opcion == 1:

            agregar_paciente()

        elif opcion == 2:

            leer_paciente()

        elif opcion == 3:

            listar_pacientes()

        elif opcion == 4:

            buscar_paciente()

        elif opcion == 5:

            modificar_paciente()

        elif opcion == 6:

            borrar_paciente()

        elif opcion == 7:

            print("Saliendo del sistema...")
            break

        else:

            print("Opción inválida.")

# --------------------------------------------------
# Ejecutar menú
menu()
```
